```python
#Importing required libraries
import pandas as pd
```

```python
#Reading the CSV file and seperating it according to the data
users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
```

```python
#Discover what is the mean age per occupation
users.groupby('occupation')['age'].mean()
```

```python
#Discover the Male ratio per occupation and sort it from the most to the least
users['gender_count'] = users['gender'].map({'M':1,'F':0})
users_sort = users.groupby('occupation')['gender_count'].mean()*100
users_sort.sort_values(ascending = False)
```

```python
#For each occupation, calculate the minimum and maximum ages
users.groupby('occupation')['age'].agg(['min', 'max'])
```

```python
#For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation', 'gender'])['age'].mean()
```

```python
#For each occupation present the percentage of women and men
users.groupby('occupation')['gender'].value_counts(normalize=True)*100
```

```python
#Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading data from CSV file
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
data.head()
```

```python
#Generating chart according to instructions
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(10,8)})
chart = sns.lineplot(x="Year", y='Marriages',data=data,label = 'Marriages')
chart = sns.lineplot(x="Year", y='Divorces' ,data=data,label = 'Divorces')
chart.set_title('Marriages and Divorces from 1867 to 2014',weight='bold')
chart.set_ylabel('Marriages and Divorces Counts(Millions)')
plt.show()
```

<h3>Question 7</h3>
<p>
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.<br>
Don't forget to label your axes!<br>
</p>

```python
#Generating bar chart
comparison = data[(data.Year == 1900) | (data.Year == 1950) | (data.Year == 2000)]
comparison_1 = comparison.drop(columns = ['Marriages', 'Divorces', 'Population']).set_index('Year')
comparison_1.plot.bar(figsize=(10,5))
plt.xlabel('Years', fontsize = 14)
plt.ylabel('No of Marriages & Divorces', fontsize = 14) 
plt.title("Number of marriages & Divorces per Capita in U.S. between 1900, 1950, and 2000", fontsize=15)
plt.legend() 
plt.show()
```
